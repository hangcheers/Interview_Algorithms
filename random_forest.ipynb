```python
import pandas as pd
from sklearn.datasets import load_breast_cancer

columns = ['mean radius', 'mean texture', 'mean perimeter', 'mean area', 'mean smoothness', 'mean compactness',
           'mean concavity', 'mean concave points', 'mean symmetry', 'mean fractal dimension', 'radius error',
           'texture error', 'perimeter error', 'area error', 'smoothness error', 'compactness error', 'concavity error',
           'concave points error', 'symmetry error', 'fractal dimension error', 'worst radius', 'worst texture',
           'worst perimeter', 'worst area', 'worst smoothness', 'worst compactness', 'worst concavity', 'worst concave points',
           'worst symmetry', 'worst fractal dimension']

dataset = load_breast_cancer()
data = pd.DataFrame(dataset['data'], columns=columns)
data['cancer'] = dataset['target']
data.head()
```

```python
from sklearn.model_selection import train_test_split

X = data.drop('cancer', axis=1)
y = data['cancer']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=2020, stratify=y)
```

scale means to change the range of the values. The range is often set at 0 to 1. 
standardize means to change the values so that the distribution std from the mean equals to 1.
{kernel density estimate}

```python
# scale the data
import numpy as np
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
X_train_scaled = ss.fit_transform(X_train)
X_test_scaled = ss.fit_transform(X_test)
#y_train = np.array(y_train)

X_train_scaled
```

```python
# build the baseline of random forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import recall_score

rfc = RandomForestClassifier()
rfc.fit(X_train_scaled, y_train)
rfc.score(X_train_scaled, y_train)
```

```python
# quantify the feature importances
feats = {}
for feature, importance in zip(data.columns, rfc.feature_importances_):
    feats[feature] = importance
feats
```

```python
importances = pd.DataFrame.from_dict(feats, orient='index').rename(columns={0: 'Gini-Importance'})
importances = importances.sort_values(by='Gini-Importance', ascending=False)
importances = importances.reset_index()
importances = importances.rename(columns={'index': 'Features'})
importances
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(font_scale = 5)
sns.set(style='whitegrid', color_codes=True, font_scale=1.7)
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)
sns.barplot(x=importances['Gini-Importance'],
           y=importances['Features'], data=importances, color='skyblue')
plt.xlabel('Importance', fontsize=25, weight='bold')
plt.ylabel('Features', fontsize=25, weight='bold')
plt.title('Feautre Importances', fontsize=25, weight='bold')
plt.show()
```

 The realtive rank of a feature used as a decision node in a tree can be used to assess the relative importance with respect to the predictability of the target variables. The higher the value, the more important is the contribution of the matching feature to the prediction function

```python
# the cumulative explained variance ratio
from sklearn.decomposition import PCA

pca_test = PCA(n_components=30)
pca_test.fit(X_train_scaled)
pca_test.explained_variance_ratio_
```

```python
sns.set(style='whitegrid')
plt.plot(np.cumsum(pca_test.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.axvline(linewidth=4, color='r', linestyle='--', x=10, ymin=0, ymax=1)
plt.show()
```

```python
evr = pca_test.explained_variance_ratio_
cvr = np.cumsum(pca_test.explained_variance_ratio_)
pca_df = pd.DataFrame()
pca_df['Explained Variance Ratio'] = evr
pca_df['Cumulative Variance Ratio'] = cvr

pca_df
```

```python
pca = PCA(n_components=10)
pca.fit(X_train_scaled)

X_train_scaled_pca = pca.transform(X_train_scaled)
X_test_scaled_pca = pca.transform(X_test_scaled)
```

We use PCA to reduce the number of components from 30 to 10. Each components is a linear combination of the original variables with corresponding weights. We can check these weights for each PCA components by creating a dataframe.

```python
pca_dims = []
for x in range(0, len(pca_df)):
    pca_dims.append('pca componets {}'.format(x))

pca_test_df = pd.DataFrame(pca_test.components_, columns=columns, index=pca_dims)
pca_test_df.head(10).T
```

```python
rfc1 = RandomForestClassifier()
rfc1.fit(X_train_scaled_pca, y_train)
rfc1.score(X_train_scaled_pca, y_train)
```

We start with RandomSearchCV to consider a wider range of values  

random search在参数空间随机采样， cv交叉验证

```python
from sklearn.model_selection import RandomizedSearchCV

n_estimators = [int(x) for x in np.linspace(start=100, stop=1000, num=10)]
max_features = ['log2', 'sqrt']
max_depth = [int(x) for x in np.linspace(start=1, stop=15, num=15)]
min_samples_split = [int(x) for x in np.linspace(start=2, stop=50, num=10)]
min_samples_leaf = [int(x) for x in np.linspace(start=2, stop=50, num=10)]
bootstrap = [True, False]

# param_dist字典类型，放入参数搜索范围
param_dist = {'n_estimators': n_estimators, 
              'max_features': max_features,
              'max_depth': max_depth,
              'min_samples_split': min_samples_split,
              'min_samples_leaf': min_samples_leaf,
              'bootstrap': bootstrap}

rf2 = RandomForestClassifier()

# n_iter = 100，训练100次，
# n_jobs = -1，使用所有的CPU进行训练，默认为1，使用1个CPU
rs = RandomizedSearchCV(rf2, param_dist, n_iter=10, cv=3, verbose=1, n_jobs=-1, random_state=0)
rs.fit(X_train_scaled_pca, y_train)

# best_params_ give the best performing model's parameters
rs.best_params_
```

```python
rs_df = pd.DataFrame(rs.cv_results_).sort_values('rank_test_score')
rs_df
```

```python
rs_df = rs_df.reset_index(drop=True)
rs_df
```

```python
rs_df = pd.DataFrame(rs.cv_results_).sort_values('rank_test_score').reset_index(drop=True)
rs_df = rs_df.drop(['mean_fit_time', 'std_fit_time', 'mean_score_time', 'std_score_time', 'params',
                   'split0_test_score', 'split1_test_score', 'split2_test_score', 'std_test_score'], axis=1)
rs_df
```

```python
# evaluate performance of models on the test data

# baseline random forest
y_pred = rfc.predict(X_test_scaled)
# baseline random forest with PCA reduced dimensionality
y_pred_pca = rfc1.predict(X_test_scaled_pca)
# baseline random forest with PCA reduced dimensionality & hyperparameter tuning
y_pred_rs = rs.best_estimator_.predict(X_test_scaled_pca)
```

confusion matrix is a performance measurement for machine learning classification  

confusion matrix is useful for measuring recall, precision, specificity, accuracy and AUC-ROC curve  

Actual 1, Predicted 1 is True Positive  ===> **True Positive** means that we predict positive and it is true.  
Actual 0, Predicted 0 is True Negative  ===> **True Negative** means that we predict negative and it is true.

Actual 0, Predicted 1 is False Positive ===> **False Positive** means that we predict positive and it is true.  
Actual 1, Predicted 0 is False Negative ===> **False Negative** means that we predict negative and it is False.

**We describe the predicted value as positive and negative, and whether our predict is right or not as true or false**    

Then we calculate **recall** and **precision**,   
recall = TP/(TP+FN) Out of all the actual 1, how much we have predicted 1 correctly.  
precision = TP/(TP+FP) Out of all the predicted 1, how much actual 1 we have


```python
from sklearn.metrics import confusion_matrix

conf_matrix_baseline = pd.DataFrame(confusion_matrix(y_test, y_pred),
                                   index = ['actual 0', 'actual 1'], columns = ['predicted 0', 'predicted 1'])
conf_matrix_baseline
```

```python
conf_matrix_baseline_pca = pd.DataFrame(confusion_matrix(y_test, y_pred_pca),
                                       index = ['actual 0', 'actual 1'], columns=['predicted 0', 'predicted 1'])
conf_matrix_baseline_pca
```

```python
conf_matrix_tuned_pca = pd.DataFrame(confusion_matrix(y_test, y_pred_rs),
                                    index = ['actual 0', 'actual 1'], columns=['predicted 0', 'predicted 1'])
conf_matrix_tuned_pca
```
